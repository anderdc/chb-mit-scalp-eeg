### Data Preparation

I realize we need to polish the data before we can even extract any features. I'd say this process is akin to going from bronze -> silver quality of data.
- data annotation
- data segmentation (use a window size and overlap)
- label generation (creating a vector of the target class that is |Segments| x 1 size)

```python
# load up data file!
filename = '../data/chb15/chb15_06.edf'
raw = mne.io.read_raw_edf(filename, preload=True)
raw.drop_channels(['--0', '--1', '--2', '--3', '--4', '--5'])   # dud channels?

# Annotate the data
'''
Note: we are only working with one file, and this file only has 1 seizure, so I am manually setting the annotations.
use this as a starting point for later in the project to have a script that can automate this process
'''
onsets = []
durations = []
descriptions = []
onsets.append(272) 
durations.append(397 - 272)  # end time - start time
descriptions.append("ictal")

annotations = mne.Annotations(onset=onsets, duration=durations, description=descriptions)

raw.set_annotations(annotations)    # I only set so it doesn't output the metadata
```

```python
# segment the data
WINDOW_SIZE = 2.5  # in seconds
OVERLAP = (0.5) * WINDOW_SIZE   # ALSO in seconds... doing it like this so I can just set as a % of windowsize

epochs = mne.make_fixed_length_epochs(raw, duration=WINDOW_SIZE, overlap=OVERLAP, preload=True)
print('total segments:', len(epochs))
```

```python
# Create label vector 
sfreq = raw.info["sfreq"]   # get sampling frequency
labels = []

for x in range(len(epochs.events[:, 0])):
    # using sampling freq, get start time and end time of a segment
    t0 = epochs.events[x, 0] / sfreq
    t1 = t0 + epochs.tmax
    # if there is ANY overlap of a seizure period and this segment, label = 1
    is_seizure = any(
        (ann["onset"] < t1) and (ann["onset"] + ann["duration"] > t0)
        for ann in raw.annotations if 'ictal' in ann["description"].lower()
    )
    labels.append(int(is_seizure))

print('total target segments:',sum(labels))   # rough check on how many segments overlap with ictal period
print(f'% of segments that are class=1 (ictal): {sum(labels)/len(labels)*100:.2f}%')
```

```python
# for setting what channels to plot for all features
channels_to_plot = [6, 12, 16]  # Change this list to plot different channels
n_channels = len(channels_to_plot)

# scatterplot function
def plot_feature(feature, x_label, y_label, title):
    fig, axes = plt.subplots(1, n_channels, figsize=(5 * n_channels, 4), sharey=True)
    
    for i, ch in enumerate(channels_to_plot):
        ax = axes[i]
        scatter = ax.scatter(
            range(len(feature[:, ch])),  # X-axis: epoch index
            feature[:, ch],              # Y-axis: mean for channel ch
            c=labels, cmap="coolwarm", alpha=0.8 # Color by label
        )
        ax.set_title(f"Channel {raw.ch_names[ch]}")
        ax.set_xlabel(x_label)
        if i == 0:
            ax.set_ylabel(y_label)
    
    fig.suptitle(title, fontsize=14)
    plt.tight_layout()
    plt.show()
```

#### 1. Mean Extraction

- this feature is the mean of values in a segment (epoch) of data. will be calculated per channel.
- this would result in a matrix of `(n epochs) x (n channels)`

List of channels:

FP1-F7, F7-T7, T7-P7, P7-O1, FP1-F3, F3-C3, C3-P3, P3-O1, FZ-CZ, CZ-PZ, PZ-OZ, FP2-F4, F4-C4, C4-P4, P4-O2, FP2-F8, F8-T8, T8-P8-0, P8-O2, P7-T7, 
T7-FT9, FT9-FT10, FT10-T8, T8-P8-1, FC1-Ref, FC2-Ref, FC5-Ref, FC6-Ref, CP1-Ref, CP2-Ref, CP5-Ref, CP6-Ref

- Upon a brief visual inspection I see no significance in the mean being a predictor of ictal states, I wasn't too keen on this feature being too informative, but a later section will delve into feature quality analysis

```python
# Extracting mean per epoch (segment), per channel
data = epochs.get_data()  # shape: (n_epochs, n_channels, n_times)
means_per_epoch = data.mean(axis=2)  # shape: (n_epochs, n_channels)
```

```python
# scatter plots to show means of every segment over time, 
plot_feature(means_per_epoch, "Epoch Index", "Mean Signal Value", "Mean per Epoch for Selected Channels")
```

#### 2. VAR Extraction

- this is the variance of a channel in an epoch.
- The resulting feature matrix is size `(n epochs) x (n channels)` 

```python
data = epochs.get_data()  # shape: (n_epochs, n_channels, n_times)
vars_per_epoch = data.var(axis=2)  # shape: (n_epochs, n_channels)
```

```python
# scatter plots
fig, axes = plt.subplots(1, n_channels, figsize=(5 * n_channels, 4), sharey=True)

for i, ch in enumerate(channels_to_plot):
    ax = axes[i]
    scatter = ax.scatter(
        range(len(vars_per_epoch[:, ch])),  # X-axis: epoch index
        vars_per_epoch[:, ch],              # Y-axis: mean for channel ch
        c=labels, cmap="coolwarm", alpha=0.8 # Color by label
    )
    ax.set_title(f"Channel {raw.ch_names[ch]}")
    ax.set_xlabel("Epoch Index")
    if i == 0:
        ax.set_ylabel("Variance")

fig.suptitle("VAR per Epoch for Selected Channels", fontsize=14)
plt.tight_layout()
plt.show()
```

#### 3. MAV Extraction

- this is the mean absolute value, basically mean but take the abs of the value first
- the resulting feature matrix will be size `(n epochs) x (n channels)`

```python
data = epochs.get_data()
mavs_per_epoch = np.absolute(data).mean(axis=2)
```

```python
# scatter plots
fig, axes = plt.subplots(1, n_channels, figsize=(5 * n_channels, 4), sharey=True)

for i, ch in enumerate(channels_to_plot):
    ax = axes[i]
    scatter = ax.scatter(
        range(len(mavs_per_epoch[:, ch])),  # X-axis: epoch index
        mavs_per_epoch[:, ch],              # Y-axis: mean for channel ch
        c=labels, cmap="coolwarm", alpha=0.8 # Color by label
    )
    ax.set_title(f"Channel {raw.ch_names[ch]}")
    ax.set_xlabel("Epoch Index")
    if i == 0:
        ax.set_ylabel("Mean Absolute Value")

fig.suptitle("MAV per Epoch for Selected Channels", fontsize=14)
plt.tight_layout()
plt.show()
```

#### 4. Skewness

- Skewness is a measure of the asymmetry of a distribution
- Skewness is used to detect outliers in a data set. It also shows where the data set is trending.
- Skewness also means that the mean, median, and mode of a data set do not share the same value and are all different.
- Calculation is: `3(mean-median)/standard deviation`

```python
data = epochs.get_data()
medians_per_epoch = np.median(data, axis = 2)
std_per_epoch = data.std(axis = 2)
skew_per_epoch = 3*(means_per_epoch - medians_per_epoch) / std_per_epoch
```

```python
fig, axes = plt.subplots(1, n_channels, figsize=(5 * n_channels, 4), sharey=True)

for i, ch in enumerate(channels_to_plot):
    ax = axes[i]
    scatter = ax.scatter(
        range(len(skew_per_epoch[:, ch])),  # X-axis: epoch index
        skew_per_epoch[:, ch],              # Y-axis: mean for channel ch
        c=labels, cmap="coolwarm", alpha=0.8 # Color by label
    )
    ax.set_title(f"Channel {raw.ch_names[ch]}")
    ax.set_xlabel("Epoch Index")
    if i == 0:
        ax.set_ylabel("Skewness")

fig.suptitle("Skew per Epoch for Selected Channels", fontsize=14)
plt.tight_layout()
plt.show()
```

#### 1. Band Power &  (2) Relative Band Power

These features extract 1 value per channel per epoch. ouput size `(n_epochs) x (n_channels)` and can be calculated on a per band basis.

- Delta (0.54 Hz)
- Theta (48 Hz)
- Alpha (813 Hz)
- Beta (1330 Hz)
- Gamma (3045+ Hz)

I will extract the band power and relative band power per band (relative power = band power / overall band power)


##### 1a. Delta Band Power & Relative Band Power (0.5-4 Hz)

```python
data = epochs.get_data()
sfreq = epochs.info['sfreq']
delta_band = (0.5, 4)

delta_band_power_per_epoch = np.array([
    bandpower(epoch, sfreq, delta_band, relative=False)
    for epoch in data
])

relative_delta_band_power_per_epoch = np.array([
    bandpower(epoch, sfreq, delta_band, relative=True)
    for epoch in data
])
```

```python
plot_feature(delta_band_power_per_epoch, "Epoch Index", "Power uV^2/Hz", "Delta Band Power per Epoch for Selected Channels")
```

```python
plot_feature(relative_delta_band_power_per_epoch, "Epoch Index", "Power uV^2/Hz", "Relative Delta Band Power per Epoch for Selected Channels")
```
